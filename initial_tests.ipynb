```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge

from sklearn.metrics import mean_squared_error

#from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df_96 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED1996_97_PP.csv")
df_97 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED1997_98_PP.csv")
df_98 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED1998_99_PP.csv")
df_99 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED1999_00_PP.csv")
df_00 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED2000_01_PP.csv")
df_01 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED2001_02_PP.csv")
df_02 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED2002_03_PP.csv")
df_03 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED2003_04_PP.csv")
df_04 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED2004_05_PP.csv")
df_05 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED2005_06_PP.csv")
df_06 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED2006_07_PP.csv")
df_07 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED2007_08_PP.csv")
df_08 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED2008_09_PP.csv")
df_09 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED2009_10_PP.csv")
df_10 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED2010_11_PP.csv")
df_11 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED2011_12_PP.csv")
df_12 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED2012_13_PP.csv")
df_13 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED2013_14_PP.csv")
df_14 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED2014_15_PP.csv")
df_15 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED2015_16_PP.csv")
df_16 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED2016_17_PP.csv")
df_17 = pd.read_csv("CollegeScorecard_Raw_Data/MERGED2017_18_PP.csv")
```

```python
frames = [
    df_96, df_97, df_98, df_99, df_00,
    df_01, df_02, df_03, df_04, df_05,
    df_06, df_07, df_08, df_09, df_10,
    df_11, df_12, df_13, df_14, df_15,
    df_16, df_17
]
data = pd.concat(frames)
```

```python
print(data.shape)
data.head()
```

```python
data_removed_nulls = data.copy()
data_removed_nulls = data_removed_nulls.dropna()
print(data_removed_nulls.shape)
data_removed_nulls.head()
##Can't remove nulls
```

```python
##Drop columns with all nulls first
data_removed_nulls = data.copy()
print(data_removed_nulls.shape)
data_removed_nulls = data_removed_nulls.dropna(axis='columns', how='all')
print(data_removed_nulls.shape)

##Drop rows with nulls for our Y column
data_removed_nulls = data_removed_nulls.dropna(subset=['MD_EARN_WNE_P10'])
print(data_removed_nulls.shape)
data_removed_nulls.head()
```

```python
##Drop columns with earning nulls first
data_removed_nulls = data.copy()
print(data_removed_nulls.shape)


##Drop rows with nulls for our Y column
data_removed_nulls = data_removed_nulls.dropna(subset=['MD_EARN_WNE_P10'])
print(data_removed_nulls.shape)

data_removed_nulls = data_removed_nulls.dropna(axis='columns', how='all')
print(data_removed_nulls.shape)

data_removed_nulls.head()
```

```python
##resolve empty fields
##remove 'Privacy suppressed'
data_removed_nulls = data_removed_nulls.fillna(0)
data_removed_nulls = data_removed_nulls[data_removed_nulls.MD_EARN_WNE_P10 != 'PrivacySuppressed']
```

```python
y = data_removed_nulls['MD_EARN_WNE_P10']
y = y.astype("float64")
X = data_removed_nulls.drop(columns=['MD_EARN_WNE_P10'])
```

```python
#convert columns into floats where possible, dropping others
to_drop = []
for i in X.columns:
    if X[i].dtype == "object":
        try:
            X[i].astype('float64')
        except:
            to_drop.append(i)
print("Columns before filtering: " + str(X.shape[1]))
X = X.drop(columns=to_drop)
print("Columns after filtering: " + str(X.shape[1]))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
X_train.head()
```

```python
### Scaling
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
### Linear Model
lm_reg = LinearRegression()
lm_reg.fit(X_train_scaled, y_train)

train_pred = lm_reg.predict(X_train_scaled)
lm_pred = lm_reg.predict(X_test_scaled)

print("Train error: " + str(mean_squared_error(y_train, train_pred)))
print("Test error: " + str(mean_squared_error(y_test, lm_pred)))
```

```python
### Ridge Regression
ridge = Ridge()
ridge.fit(X_train_scaled, y_train)

ridge_train_pred = ridge.predict(X_train_scaled)
ridge_pred = ridge.predict(X_test_scaled)

print("Train error: " + str(mean_squared_error(y_train, ridge_train_pred)))
print("Test error: " + str(mean_squared_error(y_test, ridge_pred)))
```
