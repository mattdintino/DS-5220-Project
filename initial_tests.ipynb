```python
#convert columns into floats where possible, one-hot-encode others
cats = []
for i in data_removed_nulls.columns:
    if data_removed_nulls[i].dtype == "object":
        try:
            data_removed_nulls[i] = data_removed_nulls[i].astype('float64')
        except:
            cats.append(i)
```

```python
cats
```

```python
#print("Columns before filtering: " + str(X.shape[1]))
ohe = pd.get_dummies(data_removed_nulls[['CITY', 'STABBR']])
data_removed_nulls = data_removed_nulls.drop(columns=cats)
#print("Columns after filtering: " + str(X.shape[1]))
#print("ohe: " + str(ohe.shape[1]))
ohe.head()
```

```python
data_removed_nulls = data_removed_nulls.join(ohe, how='inner')
print(data_removed_nulls.shape)
data_removed_nulls.head()
```

```python
y = data_removed_nulls['MD_EARN_WNE_P10']
y = y.astype("float64")
X = data_removed_nulls.drop(columns=['MD_EARN_WNE_P10'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
X_train.head()
```

```python
### Scaling
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
### Linear Model
lm_reg = LinearRegression()
lm_reg.fit(X_train_scaled, y_train)

train_pred = lm_reg.predict(X_train_scaled)
lm_pred = lm_reg.predict(X_test_scaled)

print("Train error: " + str(mean_squared_error(y_train, train_pred)))
print("Test error: " + str(mean_squared_error(y_test, lm_pred)))
```

```python
### Ridge Regression
ridge = Ridge()
ridge.fit(X_train_scaled, y_train)

ridge_train_pred = ridge.predict(X_train_scaled)
ridge_pred = ridge.predict(X_test_scaled)

print("Train error: " + str(mean_squared_error(y_train, ridge_train_pred)))
print("Test error: " + str(mean_squared_error(y_test, ridge_pred)))
```
