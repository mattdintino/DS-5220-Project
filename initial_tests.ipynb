```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
X_train.head()
```

X_train = X_train.as_matrix()

```python
### One Hot Encoding
enc = preprocessing.OneHotEncoder(handle_unknown='ignore')
enc.fit(X_train[cats])
ohe_train = enc.transform(X_train[cats])
ohe_test = enc.transform(X_test[cats])
```

```python
ohe_train
```

```python
### Scaling
scaler = StandardScaler()
scaler.fit(X_train.drop(columns=cats))
scaled_train = csr_matrix(scaler.transform(X_train.drop(columns=cats)))
scaled_test = csr_matrix(scaler.transform(X_test.drop(columns=cats)))
```

```python
X_mat_train = hstack((ohe_train, scaled_train))
X_mat_test = hstack((ohe_test, scaled_test))
X_mat_train
```

```python
### Linear Model
lm_reg = LinearRegression()
lm_reg.fit(X_mat_train, y_train)
```

```python
train_pred = lm_reg.predict(X_mat_train)
print("Train error: " + str(mean_squared_error(y_train, train_pred)))
```

```python
lm_pred = lm_reg.predict(X_mat_test)
print("Test error: " + str(mean_squared_error(y_test, lm_pred)))
```

```python
### Ridge Regression
ridge = Ridge()
ridge.fit(X_mat_train, y_train)

ridge_train_pred = ridge.predict(X_mat_train)
ridge_pred = ridge.predict(X_mat_test)

print("Train error: " + str(mean_squared_error(y_train, ridge_train_pred)))
print("Test error: " + str(mean_squared_error(y_test, ridge_pred)))
```

```python
#### Neural Network
mlp = MLPRegressor(activation='logistic')
mlp.fit(X_mat_train, y_train)

mlp_train_pred = mlp.predict(X_mat_train)
mlp_pred = mlp.predict(X_mat_test)

print("Train error: " + str(mean_squared_error(y_train, mlp_train_pred)))
print("Test error: " + str(mean_squared_error(y_test, mlp_pred)))
```

```python
### Ensemble Boosting
boost = GradientBoostingRegressor(n_estimators=100, max_depth=4,
                                    learning_rate=0.1, loss='huber',
                                    random_state=1)
boost.fit(X_mat_train, y_train)

boost_train_pred = boost.predict(X_mat_train)
boost_pred = boost.predict(X_mat_test)

print("Train error: " + str(mean_squared_error(y_train, boost_train_pred)))
print("Test error: " + str(mean_squared_error(y_test, boost_pred)))
```
