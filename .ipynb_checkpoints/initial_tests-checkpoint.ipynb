```python
#convert columns into floats where possible, one-hot-encode others
cats = []
for i in data.columns:
    if data[i].dtype == "object":
        try:
            data[i] = data[i].astype('float64')
            if data[i].mean() == np.nan:
                print(i)
                data = data.drop(columns=i)
            else:
                data[i] = data[i].fillna(data[i].mean())
        except:
            cats.append(i)
            data[i] = data[i].fillna("null")
    elif data[i].dtype == "float64":
        data[i] = data[i].fillna(data[i].mean())
```

```python
#print("Columns before filtering: " + str(X.shape[1]))
ohe = pd.get_dummies(data[cats])
data = data.drop(columns=cats)
#print("Columns after filtering: " + str(X.shape[1]))
#print("ohe: " + str(ohe.shape[1]))
print(ohe.shape)
ohe.head()
```

```python
print(data.shape)
data.head()
```

```python
ohe = ohe.reset_index()
data = data.reset_index()
#data.index.value_counts()
```

```python
data1 = data.join(ohe, lsuffix='_l', rsuffix='_r')
#data1 = data1.drop(columns = ['level_0', 'index_l', 'index_r'])
#data1 = pd.merge(data, ohe, left_index=True, right_index=True, how='inner');
print(data1.shape)
data1.head()
```

```python
y = data1['MN_EARN_WNE_P10']
y = y.astype("float64")
X = data1.drop(columns=['MD_EARN_WNE_P10'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
X_train.head()
```

```python
### Scaling
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
### Linear Model
lm_reg = LinearRegression()
lm_reg.fit(X_train_scaled, y_train)
```

```python
train_pred = lm_reg.predict(X_train_scaled)
print("Train error: " + str(mean_squared_error(y_train, train_pred)))
```

```python
lm_pred = lm_reg.predict(X_test_scaled)
print("Test error: " + str(mean_squared_error(y_test, lm_pred)))
```

```python
### Ridge Regression
ridge = Ridge()
ridge.fit(X_train_scaled, y_train)

ridge_train_pred = ridge.predict(X_train_scaled)
ridge_pred = ridge.predict(X_test_scaled)

print("Train error: " + str(mean_squared_error(y_train, ridge_train_pred)))
print("Test error: " + str(mean_squared_error(y_test, ridge_pred)))
```
