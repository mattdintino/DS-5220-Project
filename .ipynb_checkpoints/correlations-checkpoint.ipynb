```python
#convert columns into floats where possible, one-hot-encode others
#cats = []
for i in data.columns:
    if data[i].dtype == "object":
        try:
            data[i] = data[i].astype('float64')
            if data[i].mean() == np.nan:
                print("mean=null")
                print(i)
                data = data.drop(columns=i)
            else:
                data[i] = data[i].fillna(data[i].mean())
        except:
            #cats.append(i)
            if i not in cats:
                print("not in cats")
                print("dropping: " + str(i))
                data = data.drop(columns=i)
            else:
                data[i] = data[i].fillna("null")
    elif data[i].dtype == "float64":
        data[i] = data[i].fillna(data[i].mean())
```

```python
#drop category columns from array if column was already eliminated from dataset
remaining = []
for i in cats:
    print(i)
    try:
        data[i].head(1)
        print("Keeping: " + str(i))
        #print()
        remaining.append(i)
    except:
        print("removing: " + str(i))
cats=remaining
cats
```
