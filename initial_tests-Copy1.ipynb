```python
y = data['MN_EARN_WNE_P10']
y = y.astype("float64")
X = data.drop(columns=['MN_EARN_WNE_P10'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
X_train.head()
```

```python
y_train.describe()
```

X_train = X_train.as_matrix()
X_test = X_test.as_matrix()
y_train = y_train.as_matrix()
y_test = y_test.as_matrix()

```python
### One Hot Encoding
enc = preprocessing.OneHotEncoder(handle_unknown='ignore')
enc.fit(X_train[cats])
ohe_train = enc.transform(X_train[cats])
ohe_test = enc.transform(X_test[cats])
```

```python
ohe_train
```

```python
### Scaling
scaler = StandardScaler()
scaler.fit(X_train.drop(columns=cats))
scaled_train = csr_matrix(scaler.transform(X_train.drop(columns=cats)))
scaled_test = csr_matrix(scaler.transform(X_test.drop(columns=cats)))
```

```python
X_mat_train = hstack((ohe_train, scaled_train))
X_mat_test = hstack((ohe_test, scaled_test))
X_mat_train
```

```python
### Linear Model
lm_reg = LinearRegression()
lm_reg.fit(X_mat_train, y_train)

train_pred = lm_reg.predict(X_mat_train)
print("Train MSE: " + str(mean_squared_error(y_train, train_pred)))
print("Train RMSE: " + str(np.sqrt(mean_squared_error(y_train, train_pred))))
print("Train MAE: " + str(mean_absolute_error(y_train, train_pred)))
print("Train r2: " + str(r2_score(y_train, train_pred)))

lm_pred = lm_reg.predict(X_mat_test)
print("Test MSE: " + str(mean_squared_error(y_test, lm_pred)))
print("Test RMSE: " + str(np.sqrt(mean_squared_error(y_test, lm_pred))))
print("Test MAE: " + str(mean_absolute_error(y_test, lm_pred)))
print("Test r2: " + str(r2_score(y_test, lm_pred)))
```

```python
### Ridge Regression
ridge = Ridge()
ridge.fit(X_mat_train, y_train)

ridge_train_pred = ridge.predict(X_mat_train)
ridge_pred = ridge.predict(X_mat_test)

print("Train MSE: " + str(mean_squared_error(y_train, ridge_train_pred)))
print("Train RMSE: " + str(np.sqrt(mean_squared_error(y_train, ridge_train_pred))))
print("Train MAE: " + str(mean_absolute_error(y_train, ridge_train_pred)))
print("Train r2: " + str(r2_score(y_train, ridge_train_pred)))

print("Test MSE: " + str(mean_squared_error(y_test, ridge_pred)))
print("Test RMSE: " + str(np.sqrt(mean_squared_error(y_test, ridge_pred))))
print("Test MAE: " + str(mean_absolute_error(y_test, ridge_pred)))
print("Test r2: " + str(r2_score(y_test, ridge_pred)))
```

```python
#### Neural Network
mlp = MLPRegressor()
mlp.fit(X_mat_train, y_train)

mlp_train_pred = mlp.predict(X_mat_train)
mlp_pred = mlp.predict(X_mat_test)

print("Train MSE: " + str(mean_squared_error(y_train, mlp_train_pred)))
print("Train RMSE: " + str(np.sqrt(mean_squared_error(y_train, mlp_train_pred))))
print("Train MAE: " + str(mean_absolute_error(y_train, mlp_train_pred)))
print("Train r2: " + str(r2_score(y_train, mlp_train_pred)))

print("Test MSE: " + str(mean_squared_error(y_test, mlp_pred)))
print("Test RMSE: " + str(np.sqrt(mean_squared_error(y_test, mlp_pred))))
print("Test MAE: " + str(mean_absolute_error(y_test, mlp_pred)))
print("Test r2: " + str(r2_score(y_test, mlp_pred)))
```

```python
### Ensemble Boosting
boost = GradientBoostingRegressor()
boost.fit(X_mat_train, y_train)

boost_train_pred = boost.predict(X_mat_train)
boost_pred = boost.predict(X_mat_test)

print("Train MSE: " + str(mean_squared_error(y_train, boost_train_pred)))
print("Train RMSE: " + str(np.sqrt(mean_squared_error(y_train, boost_train_pred))))
print("Train MAE: " + str(mean_absolute_error(y_train, boost_train_pred)))
print("Train r2: " + str(r2_score(y_train, boost_train_pred)))

print("Test MSE: " + str(mean_squared_error(y_test, boost_pred)))
print("Test RMSE: " + str(np.sqrt(mean_squared_error(y_test, boost_pred))))
print("Test MAE: " + str(mean_absolute_error(y_test, boost_pred)))
print("Test r2: " + str(r2_score(y_test, boost_pred)))
```

```python
### Ada Boosting
ada = AdaBoostRegressor()
ada.fit(X_mat_train, y_train)

ada_train_pred = ada.predict(X_mat_train)
ada_pred = ada.predict(X_mat_test)

print("Train MSE: " + str(mean_squared_error(y_train, ada_train_pred)))
print("Train RMSE: " + str(np.sqrt(mean_squared_error(y_train, ada_train_pred))))
print("Train MAE: " + str(mean_absolute_error(y_train, ada_train_pred)))
print("Train r2: " + str(r2_score(y_train, ada_train_pred)))

print("Test MSE: " + str(mean_squared_error(y_test, ada_pred)))
print("Test RMSE: " + str(np.sqrt(mean_squared_error(y_test, ada_pred))))
print("Test MAE: " + str(mean_absolute_error(y_test, ada_pred)))
print("Test r2: " + str(r2_score(y_test, ada_pred)))
```

```python
### Bagging
bag = BaggingRegressor()
bag.fit(X_mat_train, y_train)

bag_train_pred = bag.predict(X_mat_train)
bag_pred = bag.predict(X_mat_test)

print("Train MSE: " + str(mean_squared_error(y_train, bag_train_pred)))
print("Train RMSE: " + str(np.sqrt(mean_squared_error(y_train, bag_train_pred))))
print("Train MAE: " + str(mean_absolute_error(y_train, bag_train_pred)))
print("Train r2: " + str(r2_score(y_train, bag_train_pred)))

print("Test MSE: " + str(mean_squared_error(y_test, bag_pred)))
print("Test RMSE: " + str(np.sqrt(mean_squared_error(y_test, bag_pred))))
print("Test MAE: " + str(mean_absolute_error(y_test, bag_pred)))
print("Test r2: " + str(r2_score(y_test, bag_pred)))
```

```python
### Grid Search
#https://scikit-learn.org/stable/auto_examples/model_selection/plot_grid_search_digits.html
grid = GridSearchCV(BaggingRegressor(), cv=10, n_jobs=20, scoring="r2",
            param_grid={})
grid.fit(X_mat_train, y_train)

print("Best parameters set found on development set:")
print()
print(grid.best_params_)
print()
print("Grid scores on development set:")
print()
means = grid.cv_results_['mean_test_score']
stds = grid.cv_results_['std_test_score']
for mean, std, params in zip(means, stds, grid.cv_results_['params']):
    print("%0.3f (+/-%0.03f) for %r"
        % (mean, std * 2, params))
```
